```python
del matches["comp"]
```

```python
del matches["notes"]
```

```python
# convert the date column into a date time.
matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches.dtypes
```

```python
# add a new column that will convert the the venue column to integers, so that we can use the variable in our model.
matches['venue_code'] = matches['venue'].astype('category').cat.codes
```

```python
matches.head()
```

```python
# convet the opponents column into numerical values.
matches['opp_code'] = matches['opponent'].astype('category').cat.codes
```

```python
matches
```

```python
# we just want the hour the matche was played, therefore we shall remove the minutes in the time.
matches['hour'] = matches['time'].str.replace(":.+","",regex=True).astype('int')
```

```python
matches.head()
```

```python
# we convert the date in the week into numerical values
matches['day_code'] = matches['date'].dt.dayofweek
```

```python
matches
```

```python
# convert result column into integars. if the teeam loses and draws the result is 0 and if the win the result is 1.
matches['results'] = (matches['result'] == 'W').astype('int')
```

```python
matches
```
