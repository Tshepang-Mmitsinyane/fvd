```python
#Now we will look at the city with the highest crime rate, ranging from Murder, Assault and rape. This will help us determine which city is the most dangerous and the safest city to live in.
```

```python
# Find the city vs murder rate

fig, ax = plt.subplots(figsize=(9, 8))
sns.barplot(x = 'City', y = 'Murder', data =df, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.suptitle ("City vs Murder rate")
plt.show()
```

```python
'''from the graph Georgia has the highest murder rate in the U.S followed by Mississpi in second and Florida in third . North Dakota has the lowest Murder rate, followed by Maine in second and 
New Hampshire in third'''
```

```python
# city vs Assault rate
fig, ax = plt.subplots(figsize=(9, 8))
sns.barplot(x = 'City', y = 'Assault', data =df, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.suptitle ("City vs Assault rate")
plt.show()
```

```python
#North Carolina has the highest assault rate in the U.S followed by Florida in second and Maryland in third . Florida has appeared for the second time in our crime analysis. The city will the lowest assualt rate is North Dakota, followed by hawaii in second and Vermont in Third
```

```python
# city vs Rape rate
fig, ax = plt.subplots(figsize=(9, 8))
sns.barplot(x = 'City', y = 'Rape', data =df, ax=ax)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)

plt.suptitle ("City vs Rape rate")
plt.show()
```

```python
# Nevada has the highest rape rate, followed by Alabama in second and Californa in third. The city/state with the lowest rape rate is North Dakota in first, followed by Maine in second and South Carolina in third.
```
